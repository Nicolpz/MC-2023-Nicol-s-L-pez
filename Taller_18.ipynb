```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
def linear_regression(x, y):
    def linear_func(x, a, b):
        return np.multiply(a, x) + b
    popt, pcov = curve_fit(linear_func, x, y)
    a, b = popt
    print(f"Linear Equation: y = {a:.2f} * x + {b:.2f}")
    plt.scatter(x, y)
    plt.plot(x, linear_func(x, *popt), 'r-')
    plt.title('Regresión lineal')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    def exponential_func(x, a, b):
        return [a * pow(2.71828, b * xi) for xi in x]
    popt, pcov = curve_fit(exponential_func, x, y)
    a, b = popt
    print(f'Regresión exponencial: y = {a:.2f}e^({b:.2f}x)')
    plt.scatter(x, y)
    plt.plot(x, exponential_func(x, *popt), 'g-')
    plt.title('Regresión exponencial')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    def power_func(x, a, b):
        return a * pow(x, b)
    popt, pcov = curve_fit(power_func, x, y)
    a, b = popt
    print(f'Regresión de potencias: y = {a:.2f}x^{b:.2f}')
    plt.scatter(x, y)
    plt.plot(x, power_func(x, *popt), 'b-')
    plt.title('Regresión de potencias')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    def growth_ratio_func(x, a, b):
        return np.multiply(a, np.divide(x, np.add(x, b)))
    popt, pcov = curve_fit(growth_ratio_func, x, y)
    a, b = popt
    print(f'Regresión de razones de crecimiento: y = {a:.2f}x / (x + {b:.2f})')
    plt.scatter(x, y)
    plt.plot(x, growth_ratio_func(x, *popt), 'm-')
    plt.title('Regresión de razones de crecimiento')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
x = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5]
y = [4.2, 6.5, 7.5, 8.0, 8.5, 8.8, 9.0, 9.5]
linear_regression(x, y)
```
