```python
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.array([[0], [1], [2], [3], [4], [5], [6], [7]])
y = np.array([3.5, 2.5, 3, 1.5, 2, 1.3, 1, 0.3])
model = LinearRegression().fit(X, y)
print("Coeficiente 'a':", model.coef_[0])
print("Coeficiente 'b':", model.intercept_)
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(X, y)
x_values = np.array([0, 7])
y_values = model.predict(x_values.reshape(-1, 1))
ax.plot(x_values, y_values, color = "pink")
plt.title("Regresi√≥n lineal por minimos cuadrados")
plt.show()
```
