```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([2, 4, 6, 8, 10, 12])
y = np.array([2.2, 3, 4.5, 6, 8.5, 12])
y_log = np.log(y)
model = LinearRegression().fit(x.reshape(-1, 1), y_log)
a1 = model.coef_[0]
a0 = np.exp(model.intercept_)
equation = f"y = {a0:.2f} * exp({a1:.2f} * x)"
x_values = np.linspace(0, 14, 100)
y_values = a0 * np.exp(a1 * x_values)
plt.scatter(x, y)
plt.plot(x_values, y_values, color="pink")
plt.title("Regresión exponencial por mínimos cuadrados")
plt.xlabel("x")
plt.ylabel("y")
plt.text(0.5, 11, equation, fontsize=12, ha="left", va="center")
plt.show()
```
