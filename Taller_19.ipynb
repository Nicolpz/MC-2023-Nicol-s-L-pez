```python
import math
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
y = [0.2, 0.5, 1.8, 3.4, 5.7, 9.0, 13.8]
mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)
sum_xy = sum([(xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y)])
sum_x_squared = sum([(xi - mean_x) ** 2 for xi in x])
sum_y_squared = sum([(yi - mean_y) ** 2 for yi in y])
sy = math.sqrt(sum_y_squared / len(y))
s_tx = math.sqrt((sum_y_squared - (sum_xy ** 2) / sum_x_squared) / (len(x) - 2))
r = sum_xy / (math.sqrt(sum_x_squared) * math.sqrt(sum_y_squared))*100
print("Desviación estándar (sy):", sy)
print("Error estándar de la estimación (sΤyΤx):", s_tx)
print("Coeficiente de correlación (r):", r)
plt.scatter(x, y, color='b', label='Datos reales')
plt.plot(x, y, color='r', label='Regresión lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```
